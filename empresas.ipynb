```python
import pandas as pd
import os
```

# CRIAR ARQUIVO DE EMPRESAS

```python
file_list = os.listdir('empresas/')
file_list
```

```python
empresas = pd.DataFrame()

for file in file_list:
    df_temp = pd.read_csv('empresas/'+ file, sep=';', header=None, dtype={0: str})
    empresas = empresas.append(df_temp, ignore_index=True)
```

```python
empresas.columns = ['CNPJ_BASICO', 'RAZAO_SOCIAL', 'NATUREZA_JUR', 'QUAL_RESP','CAP_SOCIAL', 'PORTE', 'ENTE_FED_RESP']
```

```python
empresas.head(2)
```

```python
empresas.loc[empresas["PORTE"] == 0, "PORTE"] = 'NAO INFORMADO'
empresas.loc[empresas["PORTE"] == 1, "PORTE"] = 'MICRO EMPRESA'
empresas.loc[empresas["PORTE"] == 3, "PORTE"] = 'EMPRESA DE PEQUENO PORTE'
empresas.loc[empresas["PORTE"] == 5, "PORTE"] = 'DEMAIS'
```

```python
empresas = empresas[['CNPJ_BASICO', 'RAZAO_SOCIAL', 'CAP_SOCIAL', 'PORTE']]
```

```python
empresas.to_csv('empresas.csv', index=False)
```
